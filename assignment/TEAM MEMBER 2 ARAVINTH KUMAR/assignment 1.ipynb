####  *args and **kwargs

```python
def add(*num):
    c = 0
    for i in num:
        c = c+i
    return c
```

```python
add(2,7,8,90,67)
```

```python
def info(**data):
    for key,value in data.items():
        print("the key is ",key," value is ",value)
```

```python
info(name="swathi",qualification="m.tech")
```
